# Billboard Hot 100 Export 

```python
import billboard
import datetime
import csv
```

```python
# Define the start and end dates
start_date = datetime.date(2020, 1, 1)
end_date = datetime.date(2023, 4, 30)

# Define the chart name
chart_name = "hot-100"

# Define the CSV output file name
output_file = "billboard_hot_100_2020_2023.csv"

# Create a set to store unique song entries
unique_songs = set()

# Loop through dates
current_date = start_date
while current_date <= end_date:
    print(f"Fetching chart for {current_date}")

    try:
        chart = billboard.ChartData(chart_name, current_date.strftime("%Y-%m-%d"))

        for entry in chart:
            # Create a tuple with song title and artist
            song_entry = (entry.title, entry.artist)

            # Add the tuple to the set (duplicates are ignored)
            unique_songs.add(song_entry)

        # Move to the next chart (increment by 7 days)
        current_date += datetime.timedelta(days=7)

    except billboard.BillboardNotFoundException:
        print(f"No chart data found for {current_date}")
        current_date += datetime.timedelta(days=7)
        continue

# Write the unique songs to a CSV file
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["Title", "Artist"])

    for song_entry in unique_songs:
        csv_writer.writerow(song_entry)
        

print(f"Exported unique songs to {output_file}")
```
